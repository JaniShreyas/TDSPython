```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_csv("tdsw5.csv")
df
```

```python
data = np.array(df)
data = data[:, 1:]

kmeans = KMeans(n_clusters=7, random_state=1729)
kmeans.fit(data)
```
